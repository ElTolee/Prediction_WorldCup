```python
import pandas as pd
```

```python
df_historical_data= pd.read_csv('fifa_worldcup_data.csv')
df_missing_data = pd.read_csv('fifa_worldcup_missing_data.csv')
```

```python
df_missing_data.dropna(inplace=True)
df_historical_data= pd.concat([df_historical_data,df_missing_data],ignore_index=True)
df_historical_data.drop_duplicates(inplace=True)
df_historical_data.sort_values('Año',inplace=True)
```

```python
index_delete =df_historical_data[df_historical_data['Local'].str.contains('Sweden') & df_historical_data['Visita'].str.contains('Austria')].index
df_historical_data
```

```python
df_historical_data.drop(index=index_delete,inplace=True)
df_historical_data
```

```python
df_historical_data['Resultado']= df_historical_data['Resultado'].str.replace('[^\d-]','',regex=True)

df_historical_data['Local'] = df_historical_data['Local'].str.strip()
df_historical_data['Visita'] = df_historical_data['Visita'].str.strip()


```

```python
df_historical_data[['GolesLocal','GolesVisita']]=df_historical_data['Resultado'].str.split('-',expand=True)
df_historical_data
```

```python
# df_historical_data.drop('Resultado',axis=1,inplace=True)
df_historical_data.rename(columns={'Local':'HomeTeam', 'Visita':'AwayTeam','Año':'Year','GolesLocal':'HomeGoals','GolesVisita':'AwayGoals'},inplace=True)
df_historical_data
```

```python
df_historical_data=df_historical_data.astype({'HomeGoals':int, 'AwayGoals': int})
df_historical_data.dtypes
```

```python
df_historical_data['TotalGoals']= df_historical_data['HomeGoals'] + df_historical_data['AwayGoals']
df_historical_data
```

```python
df_historical_data.to_csv('clean_fifa_worldcup_matches.csv', index= False)
```
